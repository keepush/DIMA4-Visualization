## 스타일
- plot() 함수에 다양한 키워드 인수를 전달하여 스타일을 바꿀 수 있다.

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2,
         color='red',
         linestyle='-.',  # solid(-), dashed(--), dotted(:), dashdot(-.)
         linewidth=2.5,
         marker='o',
         markeredgecolor='green',
         markerfacecolor='yellow',
         markersize=9.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]

plt.plot(a, c, 'r--', label='Model length')
plt.plot(a, d, 'b:', label = 'Data length')
plt.plot(a, c+d, 'g', label = 'Total message length')

plt.legend(loc='best', shadow=True)
plt.show()
```

## 서브플롯 그리기

### subplot() 함수
- subplot() 함수를 이용해서 하나의 그림에 여러 개의 그래프를 그릴 수 있다.
      - nrows, ncols, index 의 인수를 이용하여 서브플롯의 위치를 설정한다.

```python
x1 = np.linspace(0., 5., 50)   # 균등 분할
x2 = np.linspace(0., 2., 50)

y1 = np.cos(2 * np.pi * x1) + np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)  # nrow = 2, ncols = 1, index = 1   =>  2행 1열 그리겠다
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time(s)')
plt.ylabel('Damped')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title('2st Graph')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.tight_layout()  # 서브 플롯의 레이아웃을 자동으로 맞춰 준다. 
plt.show()
```

### subplots() 함수
- subplots() 함수는 Figure, Axes 객체를 반환한다.
- Axes 클래스: 공간이 확보된 이미지 영역 정보를 가지고 있다.
    - 주요 인자
      >- nrows, ncols: 서브플롯의 행과 열 크기 기정
      >- sharex, sharey: boolean값 (default:False), x, y 축의 값 공유 여부

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 4, sharey=True)

fig.set_size_inches((16,4))    # 튜플 형태로 데이터 전달하기 때문에 괄호 2개

axes[0].scatter(np.random.random(100), np.random.random(100), c='cyan', edgecolor='red', alpha=0.5)
axes[1].scatter(np.random.random(100), np.random.random(100), c='magenta', edgecolor='green', alpha=0.5)
axes[2].scatter(np.random.random(100), np.random.random(100), c='yellow', edgecolor='blue', alpha=0.5)
axes[3].scatter(np.random.random(100), np.random.random(100), c='white', edgecolor='black', alpha=0.5)

axes[0].set_title('subplot-0')
axes[1].set_title('subplot-1')
axes[2].set_title('subplot-2')
axes[3].set_title('subplot-3')

plt.show()
```

# 막대 그래프 그리기

## 수직 막대 그래프

```python
x = np.arange(0,20)
y = x*2

x_ticks = np.arange(len(x))  # x 축의 눈금을 표시할 index 값
plt.bar(x, y, label='Blue bar', color='blue', edgecolor='yellow', linewidth=3)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
plt.xticks(x_ticks, x_ticks+1)
plt.legend()
plt.show()
```

## 수평 막대 그래프

```python
x = np.arange(0,20)
y = x*2

y_ticks = np.arange(len(x))  # x 축의 눈금을 표시할 index 값

# 수평 막대 그래프를 그린다고 해서 x와 y 값의 위치가 바뀌는 것은 아니다   (plt.bar -> plt.barh)
plt.barh(x, y, label='Blue bar', color='blue', edgecolor='yellow', linewidth=3)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('H-Bar Chart Example')
plt.yticks(y_ticks, y_ticks+1)
plt.legend()
plt.show()
```

```python
label = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 34, 30, 25, 27]
women_means = [25, 32, 34, 20, 25]

x_ticks_index = np.arange(len(men_means))
width = 0.35  # 막대 그래프 폭의 크기
plt.bar(x_ticks_index-width/2, men_means, width=width, color='dodgerblue', label='Men')
plt.bar(x_ticks_index+width/2, women_means, width=width, color='violet', label='Women')

plt.xlabel('City')
plt.ylabel('Score')
plt.title('Score by group and gender')
plt.xticks(x_ticks_index, label)
plt.legend()
plt.show()
```

# 산점도 그리기
- 산점도(scatter plot)는 직교좌표계를 이용해 두 변수 간의 상관관계를 확인하고자 할 때 사용하는 플롯이다.

```python
x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 5]

y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker='v', color='green')
plt.scatter(x2, y3, marker='^', color='red')
plt.title('Scatter Plot Example')
plt.show()
```

```python
for color in ['blue', 'orange', 'green']:
    n = 50
    x, y = np.random.rand(2, n)   # = [0,1) 사이의 난수 2 x 50 모양으로 생성
    scale = 190.0 * np.random.rand(n) + 10  # [10, 200) 사이의 난수 50개 발생
    plt.scatter(x, y, c=color, s=scale, label=color, alpha=0.3, edgecolor='none')

plt.legend(loc=(1.04, 0.7))
plt.grid(True)
plt.show()
```

# 히스토그램 그리기
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 해당하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency table)라고 한다.
- 히스토그램은 도수분포표에서 각 구간별 데이터의 빈도수를 막대 그래프로 나타낸 것이다.

```python
y = np.random.randn(1000)  # 평균이 0, 표준편차가 1인 표준정규분포에서 난수 1000개 생성
print(np.mean(y), np.std(y))
x = np.arange(len(y))
plt.bar(x, y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y, bins=50)  # bins = 데이터를 분할할 구간 수
plt.title('Histogram')
plt.show()

plt.hist(y, cumulative=True, bins=20)
plt.title('Cumulative Histogram')
plt.show()
```

# 원 그래프 그리기

```python
import matplotlib.pyplot as plt

label = ['Python', 'Java', 'C']
# autopct: 비율(%)을 표시하는 형식 문자열을 지정한다.
plt.pie([0.3, 0.35, 0.4], labels=label, autopct='%.1f%%')
plt.title('Popular Language Ranking')
plt.legend(loc = (1,0.3))
plt.show()
```

```python
data = [0.3, 0.35, 0.4]

label = ['Python', 'Java', 'C']
colors = ['coral', 'dodgerblue', 'yellowgreen']

plt.pie(
    data,  # 조각의 크기를 결정하는 값
    labels=label,  # 조각의 이름 표시
    explode=(0.1, 0, 0),  # 조각을 중심으로부터 얼마나 떨어지게 할 것인지를 설정한다. (0.1 = 반지름의 10%만큼 벗어남)
    colors=colors,  # 조각의 색깔
    autopct='%.2f%%',  # 비율을 표시하는 형식 문자열
    shadow=True,  # 그림자 표시 여부
    startangle=90,  # 조각의 시작 부분을 각도로 표시. 해당 각도에서부터 반시계 방향으로 회전하며 조각 표시 (기본값 : 0, 3시 방향)
    counterclock=False,  # 반시계 방향으로 순서대로 표시할지 결정 (기본값: True)
    rotatelabels=True  # 조각의 이름을 회전한다. (기본값: False)
)
plt.title('Popular Language Ranking')
plt.show()
```

# 박스플롯 그리기
- 박스플롯은 데이터 집합의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(outlier)가 있는지 검사하기 수월하다.
- 최소값: 제 1사분위에서 1.5IQR(inter quntile range:사분위수범위, Q3 - Q1)을 뺀 위치
- 제 1사분위(Q1): 중앙값 기준으로 하위 50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분위(Q2): 50%의 위치에 중앙값(median)을 의미 
- 제 3사분위(Q3): 중앙값 기준으로 상위 50% 중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값: 제 3사분위에서 1.5IQR을 더한 위치
- 최대값, 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한다.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)  # 평균: 0, 표준편차: 2, 1000개
data_b = np.random.normal(-3.0, 1.5, 500)  # 평균: -3, 표준편차: 1.5, 500개
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c])
plt.ylim(-10.0, 10)  # y축 범위
plt.xlabel('Data Type')
plt.ylabel('Value')
plt.show()
```

```python
data_a = np.random.normal(0, 2.0, 1000)
print(f'min: {np.min(data_a):.2f}, median:{np.median(data_a):.2f}, max:{np.max(data_a):.2f}')
quantile = np.quantile(data_a, [0, 0.25, 0.5, 0.75, 1])
print(quantile)
IQR = quantile[3] - quantile[1]
print('IQR:', IQR)
print('MIN: ', quantile[1]-(1.5*IQR))
print('MAX: ', quantile[3]+(1.5*IQR))
```

```python
# 이상치 데이터 확인

np.random.seed(0)

spread = np.random.rand(50) * 100  # [0,100) 데이터 50개
center = np.ones(25) * 50  # 50(중앙값) 25개
flier_high = np.random.rand(10) * 100 + 100  # [0,200) 데이터 10개
flier_low = np.random.rand(10) * -100  # (-99, 0] 데이터 10개

data = np.concatenate([spread, center, flier_high, flier_low])
print(data.shape)

plt.boxplot(data, 0, 'rx')
plt.grid(True)
plt.title('Box Plot')
plt.show()
```

[실습] 매체별 광고 투자 금액에 따른 상품 판매 효과 비교
- 선 그래프로 표현한다.
- 광고 매체는 tv와 google_ads 두 가지로 한다.
- x 값은(투자 금액) 5~45 사이의 5 간격으로 만들어진 연속된 값 (tv와 google_ads 공동 사용)
- 랜덤 seed는 10으로 고정
- tv의 y 값은(상품 판매 매출액) 1-64 사이의 랜덤 정수 (오름차순 정렬)
- google_ads의 y 값은 1-74 사이의 랜덤 정수 (오름차순 정렬)

```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False #유니코드가 - 기호로 설정되는 폰트가 깨지는 것을 방지
```

```python
import random

random.seed(10)
x = list(range(5,46,5))
y1 = [random.randint(1,64) for i in range(9)]
y1.sort()
y2 = [random.randint(1,74) for i in range(9)]
y2.sort()

plt.plot(x, y1, 'bs-', label='TV') 
plt.plot(x, y2, 'ro-', label='Google_ads')
plt.title('광고 매체별 상품 판매 효과 비교')
plt.xlabel('광고 투자 금액', labelpad=10)
plt.ylabel('매출')
plt.legend()
plt.show()
```
